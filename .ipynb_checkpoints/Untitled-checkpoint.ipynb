```python
shape = np.shape(img)

for index in range(4):
    img2_copy = np.array(img[:,:,:])
    for i in range(shape[0]):
        for j in range(shape[1]):
            if not j % 4 == index or not i % 4 == index:
                img2_copy[i][j] = [0,0,0]
    plt.imsave("D:/image/plot_{}.jpg".format(index),img2_copy)
    print("Save picture D:/image/plot_{}.jpg".format(index))
```

```python
def split_img(img, pic_name):
    shape = np.shape(img)
    split = 4
    max_ = 1.
    if np.max(img) > 1 :
        max_=255.

    per =  max_/split
    print(per)

    for index in range(split):
        img2_copy = np.ones(shape) * max_
        for i in range(shape[0]):
            for j in range(shape[1]):
                base_pix = np.average(img[i][j])
                if  base_pix >= index * per and base_pix < (index+1) * per:
                    img2_copy[i][j] = img[i][j]
        img2_copy = img2_copy/max_
        plt.imsave("D:/image/{}_{}.jpg".format(pic_name,index),img2_copy)
        print("Save picture D:/image/{}_{}.jpg".format(pic_name,index))
```

```python
split_img(img, "avg")
```

```python
import random

def split_img_random(img, split_num, pic_name, bg_color=[1.,1.,1.]):
    shape = np.shape(img)

    width = np.array(range(shape[0]))
    hight = np.array(range(shape[1]))
    
    mesh_matrix = np.meshgrid(width, hight)
    pixs_w = mesh_matrix[0].flatten()
    pixs_h = mesh_matrix[1].flatten()
    
    size_pix = np.size(pixs_w)
    
    indexs = np.arange(size_pix)
    random.shuffle(indexs)
    
    max_ = 1.
    if np.max(img) > 1 :
        max_=255.

    per = size_pix//split_num

    for i in range(split_num):
        img_copy = np.ones(shape) * np.array(bg_color) * max_
        
        for j in indexs[i*per: (i+1)*per]:
            img_copy[pixs_w[j], pixs_h[j]] = img[pixs_w[j], pixs_h[j]]

        img_copy = img_copy/max_
        plt.imsave("D:/image/random_split_{}_{}_{}.jpg".format(split_num,pic_name,i),img_copy)
        print("Save picture D:/image/random_split_{}_{}_{}.jpg".format(split_num,pic_name,i))
```

```python
split_img_random(img, 10, "random_16", bg_color=[0,1,0])
```

```python
def split_img_random_alpha(img_path, split_num=4, pic_name="alpha"):
    from PIL import Image
    img_ = Image.open(img_path)
    img_ = img_.convert("RGBA")
    img = np.array(img_)
    img_.close()
    
    shape = np.shape(img)

    width = np.array(range(shape[0]))
    hight = np.array(range(shape[1]))
    
    mesh_matrix = np.meshgrid(width, hight)
    pixs_w = mesh_matrix[0].flatten()
    pixs_h = mesh_matrix[1].flatten()
    
    size_pix = np.size(pixs_w)
    
    indexs = np.arange(size_pix)
    random.shuffle(indexs)

    per = size_pix//split_num

    for i in range(split_num):
        img_copy = np.zeros(shape)
        
        for j in indexs[i*per: (i+1)*per]:
            img_copy[pixs_w[j], pixs_h[j]] = img[pixs_w[j], pixs_h[j]]
        
        img_copy = Image.fromarray(np.uint8(img_copy))
        img_copy.save("D:/image/random_split_{}_{}_{}.png".format(split_num,pic_name,i),format='png')
        img_copy.close()
        print("Save picture D:/image/random_split_{}_{}_{}.png".format(split_num,pic_name,i))
```

```python
split_img_random_alpha(img_path="D:/image/123.png", split_num=20)
```
