```python
import matplotlib.pyplot as plt
```

```python
labels = "Frogs", "Hogs", "Dogs", "Logs"
```

```python
size = [15, 30, 45, 10]
```

```python
colors = ["yellowgreen", "gold", "lightskyblue", "lightcoral"]
```

```python
explode = [0, 0.1, 0, 0]
```

```python
plt.pie(size, labels=labels, colors=colors, explode=explode, autopct="%1.1f%%", shadow=True,startangle=90)
plt.axis("equal")
plt.show()
```
