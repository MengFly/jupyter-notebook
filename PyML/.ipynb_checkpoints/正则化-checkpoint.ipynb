## 正则化

正则化最简单的理解即： 为权重添加一个惩罚项，那个权重大，就惩罚那一项多一些。

那么正则化的一个直接结果就是，添加了正则化项之后，模型的参数会趋向于变小。

### Logistic回归中的正则化

机器学习中常用的正则化是L2正则化，在Logistic中，添加了正则化项的的损失函数如下：

$J(\boldsymbol{w}) = \{ \sum_{i=1}^{m}-y^{(i)}\log(\hat y^{(i)}) - (1-y^{(i)})\log(1-\hat y^{(i)})\} + \frac{\lambda}{2}\|\boldsymbol{w}\|^2$

这里借用一下SVM中的参数C，其中 $C = \frac{1}{\lambda}$ 我们将上面的损失函数替换成 $CJ(\boldsymbol{w})$

于是损失函数就变成了：

$J(\boldsymbol{w}) = C\{ \sum_{i=1}^{m}-y^{(i)}\log(\hat y^{(i)}) - (1-y^{(i)})\log(1-\hat y^{(i)})\} + \frac{1}{2}\|\boldsymbol{w}\|^2$

减小参数C就相当于增加了正则化项，我们可以绘制出模型的参数受正则化影响的图像。

下面的图像可以看到，正则化项越大，其模型参数越小.

```python
from sklearn.linear_model import LogisticRegression
from util import load_std_data
import matplotlib.pyplot as plt
import numpy as np

X_train_std, y_train, X_test_std, y_test = load_std_data()
weights, params = [],[]

for c in range(-5, 5):
    lr = LogisticRegression(C=10**c, random_state=0)
    lr.fit(X_train_std, y_train)
    weights.append(lr.coef_[1])
    params.append(10 ** c)

weights = np.array(weights)
plt.plot(params, weights[:, 0], label="petal length")
plt.plot(params, weights[:, 1], label="petal width", linestyle="--")

plt.ylabel("weight coefficient")
plt.xlabel("C")
plt.legend(loc="upper left")
plt.xscale("log")
plt.show()
```

### 使用SVM进行数据分类

```python
from sklearn.svm import SVC
from util import plot_decision_regions

svm = SVC(kernel="linear", C=1.0, random_state=0)

svm.fit(X_train_std, y_train)

X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))

plot_decision_regions(X_combined_std, y_combined, classifier=svm, test_idx=range(105, 150))

plt.xlabel("petal length [standardized]")
plt.ylabel("petal width [standardized]")

plt.show()
```

### SVM解决非线性问题

XOR样本数据：

```python
import numpy as np
import matplotlib.pyplot as plt

X_xor = np.random.randn(200, 2)

y_xor = np.logical_xor(X_xor[:, 0] > 0, X_xor[:, 1] > 0)
y_xor = np.where(y_xor, 1, -1)

plt.scatter(X_xor[y_xor==1, 0], X_xor[y_xor==1, 1], c="b", marker="x", label="1")
plt.scatter(X_xor[y_xor==-1, 0], X_xor[y_xor==-1, 1], c="r", marker="s", label="-1")

plt.ylim(-3.0)
plt.legend()
plt.show()
```

使用SVM解决非线性分体的一个目标是怎么把原始的数据进行变换到另一个特征空间中，然后使用SVM对这个映射空间进行分类。

这里最广为使用额一个核函数是径向基函数核或者高斯核：

$$k(x^{(i)}, x^{(j)}) = \exp(-
\frac{\|x^{(i)}-x^{(j)}\|^2}{2\sigma^2})$$

这里即将样本的距离转换为相似性评分。

这里，也通常将公式写成下面这样：

$$k(x^{(i)}, x^{(j)}) = \exp(-
\gamma\|x^{(i)}-x^{(j)}\|^2)$$

```python
from sklearn.svm import SVC
from util import plot_decision_regions

svm = SVC(kernel="rbf", random_state=0, gamma=0.10, C=10.0)

svm.fit(X_xor, y_xor)

plot_decision_regions(X_xor, y_xor, classifier=svm)
plt.show()
```

#### $\gamma$ 的理解

首先，我们看到，高斯核其实是度量数据之间的距离，从而将距离转换为一个到 \[0-1\] 的相似度的数。

同时 $\gamma$ 控制着这个距离的大小，就是怎样算相似，怎样算不相似。

很明显，当 $\gamma$ 的值比较大的时候，判定的距离就比较小，而当 $\gamma$ 比较小的时候，判定的距离就比较大。

下面仍然用 xor 数据，尝试使用不同的 $\gamma$ 值进行训练即可发现其中的规律。上面的例子中已经使用了比较小的 $\gamma$ 值，下面就使用一个比较大的$\gamma$ 值

```python
svm = SVC(kernel="rbf", random_state=0, gamma=50, C=1.0)

svm.fit(X_xor, y_xor)

plot_decision_regions(X_xor, y_xor, classifier=svm)
```

这个时候就看到，由于 $\Gamma$ 的值比较大。它的判定相似度就更加紧凑，其实就是判断相似的条件更加的苛刻了，只有离点比较近的位置才会被判定为相似。
